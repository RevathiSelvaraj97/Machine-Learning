Fault steel plates classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
steel_data=pd.read_csv("faults.csv")
steel_data.head(10)
```

```python
#check for null values 
steel_data.info()
```

```python
#5 number summary of dataset
steel_data.describe()
```

## Lets split the dependent and independant features,the last 7 columns are the one-hot encoded columns of the classes, we will reduce those 7 columns into 1 

```python
#split the dependent and independent values
X_data=steel_data.iloc[:,:27]
X_data
```

```python
#punching the encoded class data into one column
Y_data=[np.argmax(x) for x in steel_data.iloc[:,27:].values]
print(Y_data)
```

```python
#Lets change the y_data labels for better understanding in visualization
labels=["pastry","Z_Scratch","K_Scratch","Stains","Dirtiness","Bumps","Other_Faults"]
y_label=[labels[x] for x in Y_data]
print(y_label)
```

```python
#Lets find the relation between independent with all depndent data
for a in range(len(X_data.columns)):
    plt.bar(y_label,X_data[X_data.columns[a]])
    plt.ylabel(X_data.columns[a])
    plt.xlabel("Classs")
    plt.show()
    
        
```

```python
#Check for Outliers
for a in range(len(X_data.columns)):
    sns.boxplot(x=X_data[X_data.columns[a]],y=y_label)
    plt.show()
    
        
```

```python
#Check for whether the dataset is balanced or not?
from collections import Counter
bars=plt.bar(Counter(np.array(y_label)).keys(),Counter(np.array(y_label)).values())
for bar in bars:
    yval=bar.get_height()
    plt.text(bar.get_x(),yval+0.05,yval)
#sns.countplot(np.array(y_label))
#for x in Counter(np.array(y_label)).keys():
#    plt.text(x,Counter(np.array(y_label)).keys())
```

```python
plt.pie(x=Counter(np.array(y_label)).values(),labels=Counter(np.array(y_label)).keys(),autopct="%.2f%%")
```

## Bar and Pie charts clearly says that the class distribution is not uniform and results imbalanced dataset, will handle them by applying over sampling. 

```python
#Rebalancing the dataset with random(under) + Smote(over) Sampling technique
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.svm import SVC

strata_over={0:500,1:550,2:550,3:500,4:500,5:500}

sm=SMOTE(sampling_strategy=strata_over,random_state=9)
X_res,Y_res=sm.fit_resample(X_data,Y_data)
print(Counter(Y_res))
X_res

```

## From independent data, we can see that each column is of different units and will scale them for processing


```python
from sklearn.preprocessing import StandardScaler
X_datax=pd.DataFrame(StandardScaler().fit_transform(X_res),columns=X_res.columns)
X_datax
```

# Feature Selection
## Finding the correlation using heatmap and select features for our model

```python
plt.figure(figsize=(30,20))
sns.heatmap(X_res.corr(),annot=True,annot_kws={'size':16})
```

## From above heatmap 
> X_MIN,X_MAX and Y_MIN,Y_MAX are very much correlated; 
> Pixel_Areas is very much correlated to X_Perimter and Sum of luminosity;
> X_Perimeter and Y_perimeter are much correlated

## from all these correlations, will select Pixel_Areas, X_min,Y_min
## Also Type of Steel was one-hot encoded so will remove TypeOfSteel_A400 from dataset and there will be no harm

```python
X=X_datax.drop(labels=["X_Perimeter","Y_Perimeter","Sum_of_Luminosity","X_Minimum","Y_Minimum","TypeOfSteel_A400"],axis=1)
X
```

```python
#Split data for train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y_res,test_size=0.25,random_state=7)
```

```python
#Model building
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=7)
classifier.fit(x_train,y_train)
```

```python
#Prediction
y_pred=classifier.predict(x_test)
```

```python
#Checking Model Accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#Confusion Matrix
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,annot_kws={'size':13})
```

```python
print("Classification Report:",classification_report(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
```
